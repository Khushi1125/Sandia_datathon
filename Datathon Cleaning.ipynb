Data Cleaning and Imports

```python
import pandas as pd
import re
```

```python
df = pd.read_csv("combined_output.csv")
```

```python
df
```

```python
def split_string_number(s):
    match = re.match(r'([^\d]+)(\d+)$', s)
    if match:
        return match.groups()
    else:
        return s, None
```

```python
df[['strings', 'sub_num']] = df['file_id'].apply(split_string_number).apply(pd.Series) 
```

```python
df
```

```python
split_cols = df['strings'].str.split(r'[_-]', expand=True)
```

```python
split_cols
```

```python
split_cols = split_cols.rename(columns = {0:'word', 1:'prime', 2:'target', 3:'is_target', 4:'idx'})
```

```python
split_cols
```

```python
split_cols = split_cols[['word', 'prime', 'target', 'is_target']]
```

```python
df = pd.concat([df, split_cols], axis=1)
```

```python
df
```

```python
df.set_index('file_id', inplace = True)
```

```python
print(df)
```

```python
df
```

```python
df.drop(columns=['strings'])
```

```python
df.to_csv('merged_final.csv', index = True)
```
